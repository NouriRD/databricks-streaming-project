```python
import requests
import json
from datetime import datetime
```

```python
print(CATALOG)
print(RAW_VOLUME_PATH)
print(GITHUB_API_URL)
```

Build Request Headers

```python
headers = {
    "Authorization": f"Bearer {GITHUB_TOKEN}",
    "Accept": "application/vnd.github+json"
}
```

Call GitHub API

```python
response = requests.get(
    GITHUB_API_URL,
    headers=headers,
    timeout=30
)

print(response.status_code)
```

Validate Response & Parse Events

```python
if response.status_code != 200:
    raise Exception(
        f"GitHub API Error: {response.status_code}"
    )

events = response.json()

print(f"Events received: {len(events)}")
```

Inspect One Event

```python
print(events[0].keys())
```

**Generate Timestamp**

```python
timestamp = datetime.utcnow().strftime(
    "%Y%m%d_%H%M%S"
)

file_name = f"events_{timestamp}.json"

print(file_name)
```

Define Output Path

```python
output_path = (
    f"{RAW_VOLUME_PATH}/github_events/{file_name}"
)

print(output_path)
```

Convert Events to JSONL

```python
json_lines = "\n".join(
    json.dumps(event)
    for event in events
)
```

Write File

```python
dbutils.fs.put(
    output_path,
    json_lines,
    overwrite=True
)
```

Verify File Creation

```python
display(
    dbutils.fs.ls(
        f"{RAW_VOLUME_PATH}/github_events"
    )
)
```
