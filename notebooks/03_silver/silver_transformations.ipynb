```python
bronze_df = spark.table(
    f"{CATALOG}.{BRONZE_SCHEMA}.github_events_bronze"
)

display(bronze_df.limit(5))
```

Create Actor Schema

```python
from pyspark.sql.types import *

actor_schema = StructType([
    StructField("id", LongType(), True),
    StructField("login", StringType(), True),
    StructField("display_login", StringType(), True),
    StructField("url", StringType(), True),
    StructField("avatar_url", StringType(), True)
])
```

Parse Actor JSON

```python
from pyspark.sql.functions import from_json, col

silver_actor_df = bronze_df.withColumn(
    "actor_parsed",
    from_json(
        col("actor"),
        actor_schema
    )
)
```

```python
display(
    silver_actor_df.select(
        "actor",
        "actor_parsed"
    )
)
```

Extract Useful Fields

```python
silver_actor_df = silver_actor_df.withColumn(
    "actor_id",
    col("actor_parsed.id")
).withColumn(
    "actor_login",
    col("actor_parsed.login")
)
```

```python
display(
    silver_actor_df.select(
        "actor_id",
        "actor_login"
    )
)
```

```python
display(
    silver_actor_df.limit(5)
)
```

Repo

```python
repo_schema = StructType([
    StructField("id", LongType(), True),
    StructField("name", StringType(), True),
    StructField("url", StringType(), True)
])
```

Parse Repo

```python
silver_df = silver_actor_df.withColumn(
    "repo_parsed",
    from_json(
        col("repo"),
        repo_schema
    )
)
```

Extract Fields

```python
silver_df = silver_df.withColumn(
    "repo_id",
    col("repo_parsed.id")
).withColumn(
    "repo_name",
    col("repo_parsed.name")
)
```

```python
display(
    silver_df.select(
        "actor_login",
        "repo_name",
        "type"
    )
)
```

Build Silver Dataset

```python
silver_final_df = silver_df.select(
    col("id").alias("event_id"),
    col("type").alias("event_type"),
    col("actor_id"),
    col("actor_login"),
    col("repo_id"),
    col("repo_name"),
    col("created_at")
)
```

```python
display(silver_final_df)
```

Create Silver Table Name

```python
silver_table = (
    f"{CATALOG}.{SILVER_SCHEMA}.github_events_silver"
)

print(silver_table)
```

Save Silver Table

```python
silver_final_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(silver_table)
```

Verify Silver

```python
spark.sql(f"""
SELECT *
FROM {silver_table}
LIMIT 20
""").show(truncate=False)
```
